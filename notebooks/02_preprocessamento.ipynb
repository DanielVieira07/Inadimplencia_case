```python
#importando bibliotecas
import pandas as pd 
```

```python
df_processed = pd.read_csv('../data/processed/df.csv')
```

```python
#Visualizando as colunas do dataframe
df_processed.columns
```

```python
#fazendo o processamento dos dados dropando as colunas que não serão utilizadas
df_processed.drop(columns = ["HistoricoCredito", "Proposito", "Investimentos", 
                                            "Emprego", "EstadoCivil", "Profissao", 
                                            "Dependentes", "FiadorTerceiros", "OutrosFinanciamentos", 
                                            "Profissao", "IDEMPREGO", "IDESTADOCIVIL", "IDFIADOR", "IDHABITACAO", "IDPROFISSAO",
                                            "IDPROPOSITO", "IDINVESTIMENTOS"], inplace=True)
```

```python
df_processed.drop(columns= ["IDHISTCRED", "ResidenciaDesde", "Habitacao", "IDOUTROSFINANC"], inplace=True)
```

```python
#Movendo o Atributo classe para o ultimo index do dataframe
colunas = df_processed.columns.tolist()
colunas.insert(len(colunas), colunas.pop(colunas.index('Status')))
df_processed = df_processed[colunas]
df_processed.head()
```

```python
#Separando os atributos previsores e a classe
previsores = df_processed.drop(columns=["Status"])
classe = df_processed["Status"]
```

```python
#Conversão dos atributos categoricos para atributos numericos.
from sklearn.preprocessing import LabelEncoder
colunas_categoricas =["EMPREGO", "ESTADOCIVIL", "FIADOR", "HABITACAO", "HISTORICO", 
               "OUTROSFINANCIAMENTOS", "PROFISSAO", "PROPOSITO", "INVESTIMENTOS"]

for col in colunas_categoricas:
    label = LabelEncoder()
    previsores[col] = label.fit_transform(previsores[col].astype(str))

previsores 
```

```python
#Salvando o dataframe transformado para utilizar nos modelos e treinar os algoritmos.
df_modelado = pd.concat([previsores, classe], axis=1)
df_modelado
```

```python
df_modelado.to_csv('../data/processed/df_modelado.csv', index=False)
```
