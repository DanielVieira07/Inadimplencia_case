```python
#importando bibliotecas 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

**Upload dos Dados**


```python
#lendo os csv
df_credito = pd.read_csv(r"C:\Users\Daniel Vieira\Desktop\inadiplencia_case\data\processed\CREDITO.csv", sep=";")
df_emprego = pd.read_csv(r"C:\Users\Daniel Vieira\Desktop\inadiplencia_case\data\processed\EMPREGO.csv", sep=";")
df_estadocivil = pd.read_csv(r"C:\Users\Daniel Vieira\Desktop\inadiplencia_case\data\processed\ESTADOCIVIL.csv", sep=";")
df_fiador = pd.read_csv(r"C:\Users\Daniel Vieira\Desktop\inadiplencia_case\data\processed\FIADOR.csv", sep=";")
df_habitacao = pd.read_csv(r"C:\Users\Daniel Vieira\Desktop\inadiplencia_case\data\processed\HABITACAO.csv", sep=";")
df_historico_credito = pd.read_csv(r"C:\Users\Daniel Vieira\Desktop\inadiplencia_case\data\processed\HISTORICO_CREDITO.csv", sep=";")
df_outrofinanciamento = pd.read_csv(r"C:\Users\Daniel Vieira\Desktop\inadiplencia_case\data\processed\OUTROSFINANC.csv", sep=";")
df_profissao = pd.read_csv(r"C:\Users\Daniel Vieira\Desktop\inadiplencia_case\data\processed\PROFISSAO.csv", sep=";")
df_proposito = pd.read_csv(r"C:\Users\Daniel Vieira\Desktop\inadiplencia_case\data\processed\PROPOSITO.csv", sep=";")
df_investimentos = pd.read_csv(r"C:\Users\Daniel Vieira\Desktop\inadiplencia_case\data\processed\INVESTIMENTOS.csv", sep=";")
#juntando os arquivos de acordo com o tributo ID de cada tabela com a de credito que tem em comum com as outras tabelas. 
df = pd.merge(df_credito, df_emprego, left_on= "Emprego", right_on="IDEMPREGO", how="left")
df = pd.merge(df, df_estadocivil, left_on= "EstadoCivil", right_on="IDESTADOCIVIL", how="left")
df = pd.merge(df, df_fiador, left_on= "FiadorTerceiros", right_on="IDFIADOR", how="left")
df = pd.merge(df, df_habitacao, left_on= "Habitacao", right_on="IDHABITACAO", how="left")
df = pd.merge(df, df_historico_credito, left_on= "HistoricoCredito", right_on="IDHISTCRED", how="left")
df = pd.merge(df, df_outrofinanciamento, left_on= "OutrosFinanciamentos", right_on="IDOUTROSFINANC", how="left")
df = pd.merge(df, df_profissao, left_on= "Profissao", right_on="IDPROFISSAO", how="left")
df = pd.merge(df, df_proposito, left_on= "Proposito", right_on="IDPROPOSITO", how="left")
df = pd.merge(df, df_investimentos, left_on = "Investimentos", right_on="IDINVESTIMENTOS", how="left")

```

**EXPLORAÇÃO DOS DADOS**

```python
#visualizar 
print(df.head())
```

```python
#Informações 
df.describe()
```

```python
#Verificando valores null 
for coluna in df.columns:
    if df[coluna].isnull().sum() > 0:
        print(f"Coluna: {coluna} - {df[coluna].isnull().sum()}")
```

```python
#Preenchendo os valores nulos das colunas pela moda, ou seja, o valor mais comum.
df = df.fillna(df.mode().iloc[0])
```

```python
#Tratando o valor IDCREDITO em ordem crescente
df = df.sort_values(by=["IDCREDITO"])
```

```python
#Verificando a situação dos clientes atraves dos status, e percerbemos que 0 são os clientes inadimplêntes e 1 são adimplentes.
status = df.groupby("Status")["IDCREDITO"].count().reset_index()
status.columns = ["Status", "Quantidade"]
status
```

```python
sns.barplot(x="Status", y="Quantidade", data=status)
```

Como podemos verificar temos que 70% dos clientes que tiveram o credito aprovado são adimplentes e 30% desses clientes são inadimplêntes.

```python
#Iremos entender o proposito do cliente para adquerir os creditos, os propositos podem ser importantes para saber se eles podem ou nao adquirir os creditos.
prop = df.groupby("PROPOSITO")["IDCREDITO"].count().reset_index()
prop.columns = ["PROPOSITO", "Quantidade"]
prop
```

```python
#Vamos verificar a idade dos clientes para saber se eles tem mais ou menos chance de adquirir os creditos.
duracao = df.groupby("Duracao")["IDCREDITO"].count().reset_index()
duracao.columns = ["Duracao", "Quantidade"]
duracao
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Quantidade", y="Duracao", data=duracao)
plt.title("Quantidade de Creditos disponibilizados por duracao")
plt.xlabel("Quantidade")
plt.ylabel("Duracao")
plt.show()
```

```python
sns.barplot(x="Quantidade", y="PROPOSITO", data=prop)
        
```

Podemos verificar que para reformas em geral, um veiculo novo e moveis estão em top tier dos propositos para adquerirem creditos. 


```python
#Vamos verificar a idade dos clientes para saber se eles tem mais ou menos chance de adquirir os creditos.
idade = df.groupby("Idade")["IDCREDITO"].count().reset_index()
idade.columns = ["Idade", "Quantidade"]
idade
```

```python
sns.set_theme(style="darkgrid")
sns.lmplot(x="Quantidade", y="Idade", data=idade)
```

```python
#Vamos verificar os investimentos 
sns.displot(df, y="INVESTIMENTOS")
```

```python
#Vamos verificar a corelação dos inadiplentes e adinplentes com o estado civil dos clientes
correlacao = df.groupby(["Status", "ESTADOCIVIL"])["IDCREDITO"].count().reset_index()
correlacao.columns = ["Status", "EstadoCivil", "Quantidade"]
correlacao
```

```python
#Visualizar
plt.figure(figsize=(10,6)) 
sns.barplot(x="EstadoCivil", y="Quantidade", hue="Status", data=correlacao)

plt.title("Quantidade de Creditos disponibilizados por cliente com Estado Civil e Status adimplente ou inadimplente")
plt.xlabel("EstadoCivil")
plt.ylabel("Quantidade")
plt.legend(title="Status (0 = inadimplente, 1 = adimplente)")
plt.tight_layout()
plt.show()
```
